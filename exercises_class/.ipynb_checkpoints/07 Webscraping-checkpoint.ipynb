```python
import warnings
warnings.filterwarnings('ignore')
import bs4 #beautiful soup
import requests
```

```python
r = requests.get('https://www.dr.dk/radio/programmer?side=1')
r.raise_for_status()
soup = bs4.BeautifulSoup(r.text, 'html.parser')

# find max pages of radio programs
pagination = soup.select('.pagination')
max_pages = pagination[0].get('data-pages')

for i in range(1,int(max_pages)+1):
    r = requests.get('https://www.dr.dk/radio/programmer?side={}'.format(i))
    r.raise_for_status()
    soup = bs4.BeautifulSoup(r.text, 'html.parser')
    shows_title = soup.select('div[class=spot-content] > h3')
    for title in shows_title:
        print(title.getText())
```
