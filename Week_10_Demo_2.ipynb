```python
import pandas as pd
import datetime

data_raw = pd.read_csv('./files/corona/Test_pos_over_time.csv', sep=';', thousands='.', decimal=',')
data = data_raw[:-5]
#print(data.head())

days = []
for date in data['Date']:
    timestamp = pd.Timestamp(date)
    #print(timestamp.day_name())
    #days.append({'day': timestamp.day_name()})
    days.append({'day': timestamp.dayofweek})
    
days_df = pd.DataFrame(days)
#print(days_df.head())

data.join(days_df)
#result = pd.concat([data, days_df], axis=1, join='inner')
result = pd.concat([data, days_df], axis=1).reindex(data.index)
result
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split

# decision tree

y = result['day']
x_features = ['NewPositive', 'PrevPos', 'Tested']
#y = result['NewPositive']
#x_features = ['Tested', 'day']
X = result[x_features]

corona_model = DecisionTreeRegressor(random_state=1)
corona_model.fit(X, y)
#print(corona_model.predict(X.head()))

predicted_corona = corona_model.predict(X)
print(list(predicted_corona))
#print(mean_absolute_error(y, predicted_corona))

# split data into training and validation data, for both features and target
# The split is based on a random number generator. Supplying a numeric value to
# the random_state argument guarantees we get the same split every time we
# run this script.
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
# Define model
corona_model_v2 = DecisionTreeRegressor()
# Fit model
corona_model_v2.fit(train_X, train_y)

# get predicted prices on validation data
val_predictions = corona_model_v2.predict(val_X)
#print(val_predictions)
#print(mean_absolute_error(val_y, val_predictions))
```
