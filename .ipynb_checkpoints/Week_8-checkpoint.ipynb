```python
import sqlalchemy as s_a
import mysql.connector as mysql

import pandas as pd

'''

Starting mysql server in docker:

winpty docker exec -it mysql_notebooks bash
or
docker exec -it mysql_notebooks bash

then:
mysql -u root -proot

'''

## connecting to the database using sqlalchemy
SQLALCHEMY_DATABASE_URL = "mysql+mysqlconnector://root:root@db"
engine = s_a.create_engine(SQLALCHEMY_DATABASE_URL)
print(engine)
```

```python
# creating a database
engine.execute("CREATE DATABASE IF NOT EXISTS week8")

# use new database
engine.execute("USE week8")
ResultProxy = engine.execute("SHOW TABLES")
ResultSet = ResultProxy.fetchall()
ResultSet[:]
```

```python
# load data to panda dataframe
df = pd.read_csv('files/prog_book.csv')
df.head()
```

```python
# create data from dataframe to mysql
#df.to_sql('books', schema='week8', con=engine, if_exists='replace', index = True, index_label = 'id')
df.to_sql('books', schema='week8', con=engine, if_exists='replace', index = False)
ResultProxy = engine.execute("SHOW COLUMNS FROM books")
ResultSet = ResultProxy.fetchall()
ResultSet[:]
```

```python
# get some data
connection = engine.connect()
ResultProxy = connection.execute("SELECT Book_title, Price FROM books")
#ResultProxy = connection.execute("SELECT id FROM week8.books")
ResultSet = ResultProxy.fetchall()
ResultSet[:]
```
