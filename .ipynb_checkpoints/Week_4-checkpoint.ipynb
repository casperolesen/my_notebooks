1. Open the file './befkbhalderstatkode.csv'
2. Turn the csv file into a numpy ndarray with np.genfromtxt(filename, delimiter=',', dtype=np.uint, skip_header=1)

```python
import numpy as np
filename = './files/befkbhalderstatkode.csv'
dd = np.genfromtxt(filename, delimiter=',', dtype=np.uint, skip_header=1)
```

3. Find out how many people lived in each of the 11 areas in 2015

```python
neighb = {1: 'Indre By', 2: 'Østerbro', 3: 'Nørrebro', 4: 'Vesterbro/Kgs. Enghave', 
       5: 'Valby', 6: 'Vanløse', 7: 'Brønshøj-Husum', 8: 'Bispebjerg', 9: 'Amager Øst', 
       10: 'Amager Vest', 99: 'Udenfor'}

def people_in_areas():
    res = {}
    for area in neighb:
        mask = (dd[:,0] == 2015) & (dd[:,1] == area)
        sum_people = np.sum(dd[mask][:,4])
        res[neighb[area]] = sum_people
    
    return res
people_in_areas()
```

4. Make a bar plot to show the size of each city area from the smallest to the largest

```python
%matplotlib notebook
import matplotlib.pyplot as plt

# data
populations = people_in_areas()
people_sorted = sorted(populations.values())
areas_sorted = sorted(populations, key=populations.get)
#areas = list(populations.keys())
#people = list(populations.values())

# plotting
plt.xlabel('areas')
plt.ylabel('people')
plt.bar(areas_sorted, people_sorted)
plt.xticks(rotation=45, horizontalalignment='right',fontweight='light')
```

5. Create a boolean mask to find out how many people above 65 years lived in Copenhagen in 2015

```python
mask = (dd[:,0] == 2015) & (dd[:,2] > 65)
sum_people = np.sum(dd[mask][:,4])
print('People above 65 years in Copenhagen in 2015: {}'.format(sum_people))
```

6. How many of those were from the other nordic countries (not dk)

```python
other_nordic_countries = {5101: 'Grønland',
                          5104: 'Finland',
                          5105: 'Island, ligeret dansk',
                          5106: 'Island',
                          5110: 'Norge',
                          5120: 'Sverige',
                          5901: 'Færøerne uoplyst',
                          5902: 'Færøerne'}

people_sum = 0
for country in other_nordic_countries:
    mask = (dd[:,0] == 2015) & (dd[:,2] > 65) & (dd[:,3] == country)
    people_sum += np.sum(dd[mask][:,4])
        
print('People above 65 years in Copenhagen in 2015 from other nordic countries than DK: {}'.format(int(people_sum)))
```

7. Make a line plot showing the changes of number of people in vesterbro and østerbro from 1992 to 2015

```python
%matplotlib notebook

years = list(range(1992, 2016))
vesterbro_people_years = {}
oesterbro_people_years = {}

# data
for year in years:
    v_mask = (dd[:,0] == year) & (dd[:,1] == 4)
    o_mask = (dd[:,0] == year) & (dd[:,1] == 2)
    v_sum = np.sum(dd[v_mask][:,4])
    o_sum = np.sum(dd[o_mask][:,4])
    vesterbro_people_years[year] = v_sum
    oesterbro_people_years[year] = o_sum
print(vesterbro_people_years)
print(oesterbro_people_years)

# plotting
vesterbro_x = list(vesterbro_people_years.keys())
vesterbro_y = list(vesterbro_people_years.values())

oesterbro_x = list(oesterbro_people_years.keys())
oesterbro_y = list(oesterbro_people_years.values())

plt.plot(vesterbro_x, vesterbro_y, label = 'Vesterbro')
plt.plot(oesterbro_x, oesterbro_y, label = 'Østerbro')

plt.legend()

plt.axis([1992, 2016, 40000, 80000]) #axis(x-min, x-max, y-min, y-max)


    
```
