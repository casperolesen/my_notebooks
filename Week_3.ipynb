```python
%matplotlib notebook
import matplotlib.pyplot as plt

from my_modules.Week_3 import student_generator

data = student_generator.get_students_from_csv()
names = [student.name for student in data]
grades = [student.get_avg_grade() for student in data]
plt.bar(names, grades, width=0.5, align='center')
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
prog_dict = {}

for student in data:
    if student.get_study_progression() in prog_dict:
        prog_dict[student.get_study_progression()] += 1
    else:
        prog_dict[student.get_study_progression()] = 1
        
#prog = [print(student.get_study_progression()) for student in data]

max_y_key = max(prog_dict, key=prog_dict.get) #get key to the max value of a dictionary
#print(max_y_key)
max_y_val = prog_dict[max_y_key]
#print(max_y_val)

grades = list(prog_dict.keys())
students = list(prog_dict.values())
plt.bar(grades, students, width=0.1, align='center') # bar(x-vals, y-vals, bar width, align bar relative to x-val on x-axis)

plt.axis([0-0.1, 1+0.1, 0, max_y_val+5]) #axis(x-min, x-max, y-min, y-max)
```
